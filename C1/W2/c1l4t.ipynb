```python
import numpy as np
import matplotlib.pyplot as plt
from lab_utils_multi import zscore_normalize_features, run_gradient_descent_feng

plt.rc("font", family='MicroSoft YaHei', weight="bold")
np.set_printoptions(precision=2)
```

```python
x = np.arange(0, 20, 1)
y = 1 + x ** 2
```

```python
X = x.reshape(-1, 1)
model_w, model_b = run_gradient_descent_feng(X, y, iterations=1000, alpha=1e-2)
```

```python
y_pre = np.dot(X, model_w) + model_b
# print(f"X={X},model_w={model_w},model_b={model_b}")
# print(f"y_pre={y_pre},shape={y_pre.shape}")
```

```python
plt.scatter(x, y, marker='x', c='r', label='实际值')
plt.title("未进行特征工程")
plt.plot(x, y_pre, label="预测值")
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
x = np.arange(0, 20, 1)
y = 1 + x ** 2
X = x ** 2
```

```python
X = X.reshape(-1, 1)
model_w, model_b = run_gradient_descent_feng(X, y, iterations=100000, alpha=1e-5)

plt.scatter(x, y, marker='x', c='r', label="实际值")
plt.title("添加x**2特征")
plt.plot(x, np.dot(X, model_w) + model_b, label="预测值")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
x = np.arange(0, 20, 1)
y = x ** 2

X = np.c_[x, x ** 2, x ** 3]

print(f"x={x},shape={x.shape}")
print(f"X={X},shape={X.shape}")
```

```python
model_w, model_b = run_gradient_descent_feng(X, y, iterations=10000, alpha=1e-7)

plt.scatter(x, y, marker='x', c='r', label='真实值')
plt.title("x x*2 x**3特征")
plt.plot(x, X @ model_w + model_b, label='预测值')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x = np.arange(0, 20, 1)
y = x ** 2

X = np.c_[x, x ** 2, x ** 3]
X_features = ['x', 'x^2', 'x^3']
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 3), sharey=True)

for i in range(len(ax)):
    ax[i].scatter(X[:, i], y)
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel("y")
plt.show()
```

```python
x = np.arange(0, 20, 1)
X = np.c_[x, x ** 2, x ** 3]
print(f"Raw X中每列的峰值到峰值：{np.ptp(X, axis=0)}")

X = zscore_normalize_features(X)
print(f"标准化后的X中的每列的峰值到峰值的范围：{np.ptp(X, axis=0)}")
```

```python
x = np.arange(0, 20, 1)
y = x ** 2

X = np.c_[x, x ** 2, x ** 3]
X = zscore_normalize_features(X)
model_w, model_b = run_gradient_descent_feng(X, y, iterations=100000, alpha=1e-1)

plt.scatter(x, y, marker='x', c='r', label='实际值')
plt.title("标准化的x，x^2,x^3特征")
plt.plot(x, X @ model_w + model_b, label='预测值')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
