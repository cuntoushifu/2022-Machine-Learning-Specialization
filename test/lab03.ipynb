```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('./deeplearning.mplstyle')
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300, 500], dtype=float)
x_train, y_train, x_train.shape, y_train.shape
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("房价")
plt.ylabel('价格（千刀）')
plt.xlabel("面积（平方英尺）")
plt.show()
```

```python
w, b = 100, 100
```

```python
def compute_model_output(x, w, b):
    f_wb = np.zeros(x.shape[0])
    for i in range(x.shape[0]):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
tmp_f_wb=compute_model_output(x_train,w,b)

plt.plot(x_train,tmp_f_wb,c='b',label='我们的预测')

plt.scatter(x_train,y_train,marker='x',c='r',label='实际的房价')
plt.plot(x_train,compute_model_output(x_train,300.0,100.0),c='g',label='我们的预测2')
plt.plot(x_train,compute_model_output(x_train,200.0,100.0),c='#34424b',label='我们的预测3')
plt.plot(x_train,compute_model_output(x_train,10.0,10.0),c='#6633FF',label='我们的预测4')


plt.title('房价')
plt.ylabel('价格(千刀)')
plt.xlabel('面积(平方英尺)')
plt.legend()
plt.show()
```
